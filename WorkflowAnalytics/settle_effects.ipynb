```
import matplotlib.pyplot as plt
import numpy as np

before = spark.read.format('csv').options(header='true', inferSchema='true').load('/FileStore/tables/officer_before.csv')
after = spark.read.format('csv').options(header='true', inferSchema='true').load('/FileStore/tables/officer_after.csv')
df1 = spark.table("officer_before")
df2 = spark.table("officer_after")

display(df1)
```

```
comp = df1.select('num_allegations')
display(comp)
```

```
comp2 = df2.select('num_allegations')
display(comp2)
```

```
data = [comp.collect(),comp2.collect()]
plt.boxplot(data,labels=['Before','After'])
plt.title('Number of Complaints Before and After Being Involved in a Settlement')
plt.ylabel('Number of Allegations')
plt.ylim(0,80)
display(plt.show())
```
