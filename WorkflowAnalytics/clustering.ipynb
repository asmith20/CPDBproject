```
from pyspark.ml.clustering import KMeans

data = spark.read.format('csv').options(header='true', inferSchema='true').load('/FileStore/tables/officer_before.csv')
data = spark.table("officer_before")
```

```

columns = [
 'rank1',
 'race1',
 'gender1',
 'num_allegations',
 'settlement_age',
 'sum_Attendance Recognition Award',
 'sum_Complimentary Letter',
 'sum_Department Commendation',
 'sum_Deployment Operations Center Award',
 'sum_Honorable Mention',
 'sum_Joint Operations Award',
 'sum_Life Saving Award',
 'sum_Military Deployment Award',
 'sum_Military Service Award',
 'sum_Outside Governmental Agency Award',
 'sum_Police Blue Shield Award',
 'sum_Police Officer Of The Month Award',
 'sum_Problem Solving Award',
 'sum_Special Commendation',
 "sum_Superintendent'S Honorable Mention",
 'sum_Superintendents Award Of Merit',
 'sum_Superintendents Award Of Valor',
 'sum_Traffic Stop Of The Month Award',
 'sum_Unit Meritorious Performance Award']

df = data[columns]
display(df)
```

```
import pyspark.sql.functions as func
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
import matplotlib.pyplot as plt
import numpy as np

df = df.withColumn('awards',df['sum_Attendance Recognition Award']+
 df['sum_Complimentary Letter']+
 df['sum_Department Commendation']+
 df['sum_Deployment Operations Center Award']+
 df['sum_Honorable Mention']+
 df['sum_Joint Operations Award']+
 df['sum_Life Saving Award']+
 df['sum_Military Deployment Award']+
 df['sum_Military Service Award']+
 df['sum_Outside Governmental Agency Award']+
 df['sum_Police Blue Shield Award']+
 df['sum_Police Officer Of The Month Award']+
 df['sum_Problem Solving Award']+
 df['sum_Special Commendation']+
 df["sum_Superintendent'S Honorable Mention"]+
 df['sum_Superintendents Award Of Merit']+
 df['sum_Superintendents Award Of Valor']+
 df['sum_Traffic Stop Of The Month Award']+
 df['sum_Unit Meritorious Performance Award'])

columns = [
 'rank1',
 'race1',
 'gender1',
 'num_allegations',
 'settlement_age',
 'awards']
df = df[columns]
display(df)
```

```
assembler = VectorAssembler(inputCols=columns, outputCol="features")
df = assembler.transform(df)

err = []

for a in range(2,13):
  kmeans = KMeans(k=a, seed=1)
  model = kmeans.fit(df)
  err.append(model.computeCost(df))
  
plt.plot(range(2,13),np.array(err))
display(plt.show())

```

```
kmeans = KMeans(k=6, seed=1)
model = kmeans.fit(df)

display(model, df)
```
